```python
df = pd.read_csv('PLD_seqs.csv')
df.head()
```

```python
def frac_X(seq, x):
    """Return fraction of sequence for arbitrary amino acid X."""
    return seq.count(x) / len(seq)
def frac_group(seq, group):
    """Return fraction of sequence for arbitrary group of amino acids where the group is a string of amino acid symbols."""
    count = 0
    for sym in group:
        count += seq.count(sym)
    return count / len(seq)
def frac_aromatic(seq):
    return frac_group(seq, 'FYW')
```

```python
frac_FYW = []
for i, row in df.iterrows():
    seq = df['PRDaa'][i]
    frac_FYW.append(frac_aromatic(seq))
```

```python
s1 = pd.Series(frac_FYW, name='FractionAromatic')
df = df.assign(FractionAromatic=s1.values)
df.head()
```

```python
df.to_csv("HG38_PDL_aromaticFrac.csv", index = False)
```
