```r


encodeAromaticSeq <- function(sequence) {
  
  
  aromatic <- c('F','W','Y')
  numberstring_split <- strsplit(sequence, "")[[1]]
  l=c()
  for (residue in numberstring_split){
    if (residue %in% aromatic){
      elem=1}
    else{
      elem=0
    }
    l<-c(l,elem)
    }
    
  return(array(l))
  }
```

```r

find_IDR_block <- function(original_sequence,windowsize) {
    
    
    original_sequence=gsub("[\n]", "", original_sequence)
    len_orig_seq=nchar(original_sequence)

    
    encoded_seq=encodeAromaticSeq(original_sequence)
    total_arom_number=sum(encoded_seq== 1)


    l=c()
    mean_interspacing=c()
    number_arom=c()  
    indexes=c()
    
    for (residue in seq(1, length(encoded_seq)-windowsize+1, by=1)){
        

        sequence=(encoded_seq[residue:(residue+windowsize-1)])
        
        ##find observed interarrivals:
        interarrivals=diff(which(sequence==1))


        if (length(interarrivals) == 0) {
          interarrivals=windowsize
        }


    #rate parameter (mean occurrence of aromatic residues)

    lambda=sum(sequence== 1)/length(sequence)


    #KS test to check if the inter-arrival times follow a geometrical distribution
    p_val=ks.test(interarrivals,"pgeom",prob=lambda)$p.value


    l<-c(l,p_val)

    mean_interspacing=c(mean_interspacing,mean(interarrivals))
    number_arom=c(number_arom,sum(sequence== 1))  

    }


    l[is.na(l)] = 1


    num=which.min(l)
    final_mean_interspacing=(mean_interspacing[num])
    final_number_arom=(number_arom[num])  



    new_sequence=(encoded_seq[num:(num+windowsize-1)])

    new_sequence2=substr(original_sequence,num,(num+windowsize-1))


    #define the threshold  


    thre=mean(l)/2  

    #print("threshold")  
    #print(thre)  


    thre_cut=which(as.vector(l) < thre)

    
    
    if (length(thre_cut)==0){
      #print("minima could not extended by 3 points")
    }
  

    #split consequative values
    a=split(thre_cut, cumsum(c(1, diff(thre_cut) != 1))) 

    for (residue in a){

        if ((num  %in%  residue) & (length(residue)>=2)){

            array_selected=residue
            #print("region IDR block")
            start=min(array_selected)
            end=max(array_selected)+windowsize-1
            qIDR_seq=(encoded_seq[start:end])
            total_arom_number_qIDR=sum(qIDR_seq== 1)
            space_len_qIDR=mean(diff(which(qIDR_seq==1)))

            return(c(start,end,end-start+1,min(l),final_mean_interspacing,final_number_arom,total_arom_number,len_orig_seq,total_arom_number_qIDR,space_len_qIDR))
         }   
    }  


}

```

```r
#Run Quasi IDR Finder on sequences larger than 100 aas
outcome_IDP_predictor <- list()

  for(i in 1:length(names(seq_longerthan100))){
     key=names(seq_longerthan100)[i]
     sequence=as.character(subset(data, ID==key, select="sequence"))
     outcome_IDP_predictor[[key]]=(find_IDR_block(sequence,100))
 }
```
